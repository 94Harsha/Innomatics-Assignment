User Login check


```python
#login credentials
crct_username="admin"
crct_password="1234"

#Authentication check
username=input("Enter your username: ")
password=input("Enter your password: ")

#if-else conditions
if username==crct_username and password==crct_password:
  print("Login Successful")
else:
  print("Invalid Credentials")

```

Pass/Fail Analyzer

```python
#Students marks
marks=[45,78,90,33,60]

#intialization
pass_count=0
fail_count=0

#conditions
for mark in marks:
  if mark>=50:
    pass_count+=1
  else:
    fail_count+=1
#total pass students
total_pass=pass_count
#total fail students
total_fail=fail_count

#final result
print("Total Pass Students:",total_pass)
print("Total Fail Students:",total_fail)
```

Simple Data Cleaner


```python
#given usernames
names=["Alice","bob","CHARLIE"]

cleaned_names=[]

for name in names:
  #remove extra space
  name=name.strip()
  #convert all names to lowercase
  name=name.lower()
  cleaned_names.append(name)

  #print cleaned list
  print("Cleaned Names:",cleaned_names)

```

Message length analyzer

```python
#given message
messages = ["Hi", "Welcome to the platform", "OK"]

for message in messages:
  #length of each msg
  message_length=len(message)
  print("Message:",message)
  print("Length:",message_length)

  #Flag msg longer than 10 characters
  if message_length>10:
    print("Message is longer than 10 characters")
```

Error Message Detector

```python
#given logs
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

#initialization
error_count=0

#condition
for log in logs:
  if log=="ERROR":
    error_count+=1

print("Total Errors:",error_count)
```
